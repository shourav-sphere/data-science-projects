```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset from the given URL
titanic_url = 'https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv'
titanic_df = pd.read_csv(titanic_url)
```

```python
titanic_df.head()
```

```python
# Create a new column for the number of relatives
titanic_df['Number of Relatives'] = titanic_df['SibSp'] + titanic_df['Parch']
```

```python
titanic_df.head()
```

```python
# Filter the dataframe for passengers who survived
survived_passengers = titanic_df[titanic_df['Survived'] == 1]
```

```python
# Create a scatter plot for survived passengers
plt.figure(figsize=(8, 6))
plt.scatter(survived_passengers['Number of Relatives'], survived_passengers['Age'], alpha=0.6, color='blue')
plt.title('Scatter Plot of Survived Passengers (Only Alive Passengers)')
plt.xlabel('Number of Relatives')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```
